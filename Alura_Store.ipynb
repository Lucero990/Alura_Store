```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## 1. Ingreso total por tienda

```python
#En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

ingreso_tienda = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

print("Ingreso total de la Tienda 1:" , ingreso_tienda)
print("Ingreso total de la Tienda 2:" ,ingreso_tienda2)
print("Ingreso total de la Tienda 3:" ,ingreso_tienda3)
print("Ingreso total de la Tienda 4:" ,ingreso_tienda4)
```

## 2. Ventas por categoría

```python
ventas_tienda = tienda["Categoría del Producto"].value_counts()
print("Tienda 1")
print(ventas_tienda)

ventas_tienda2 = tienda2["Categoría del Producto"].value_counts()
print("Tienda 2")
print(ventas_tienda2)

ventas_tienda3 = tienda3["Categoría del Producto"].value_counts()
print("Tienda 3")
print(ventas_tienda3)

ventas_tienda4 = tienda4["Categoría del Producto"].value_counts()
print("Tienda 4")
print(ventas_tienda4)
```

## 3. Calificación promedio de la tienda

```python
promedio_tienda = tienda["Calificación"].mean()
print("Promedio calificación Tienda 1:", promedio_tienda)

promedio_tienda2 = tienda2["Calificación"].mean()
print("Promedio calificación Tienda 2:", promedio_tienda2)

promedio_tienda3 = tienda3["Calificación"].mean()
print("Promedio calificación Tienda 3:", promedio_tienda3)

promedio_tienda4 = tienda4["Calificación"].mean()
print("Promedio calificación Tienda 4:", promedio_tienda4)


```

## 5. Envío promedio por tienda

```python
envio_t1 = tienda['Costo de envío'].mean()
print(f"Costo de envio de la tienda 1 es: {envio_t1:.2f}")


envio_t2 = tienda2['Costo de envío'].mean()
print(f"Costo de envio de la tienda 2:{envio_t2:.2f}")


envio_t3 = tienda3['Costo de envío'].mean()
print(f"Costo de envio de la tienda 3:{envio_t3:.2f}")


envio_t4 = tienda4['Costo de envío'].mean()
print(f"Costo de envio de la tienda 4: {envio_t4:.2f}")
```

## Creacion de graficos

```python


# 1. Creamos una lista con tus DataFrames y nombres para iterar
mis_tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# 2. Configuramos el lienzo para 4 gráficos (2 filas, 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Productos Destacados: Máximos y Mínimos por Tienda', fontsize=20, fontweight='bold')

for i, df in enumerate(mis_tiendas):
    # Calculamos el conteo de ventas
    conteo = df["Producto"].value_counts()

    # Identificamos TODOS los máximos y mínimos (por si hay empates)
    extremos = conteo[(conteo == conteo.max()) | (conteo == conteo.min())].sort_values(ascending=False)

    # Seleccionamos el cuadrante correspondiente
    ax = axes.flat[i]

    # Creamos el gráfico (Verde para los más vendidos, Rojo para los menos)
    colores = ['#2ecc71' if v == conteo.max() else '#e74c3c' for v in extremos.values]
    sns.barplot(x=extremos.values, y=extremos.index, ax=ax, palette=colores)

    ax.set_title(f'Análisis: {nombres_tiendas[i]}', fontsize=14)
    ax.set_xlabel('Unidades Vendidas')
    ax.set_ylabel('')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
